````python
### This cell is to calibrate and undistort the camera via images in \camera_cal ###
import glob
# Define path to create output images in
path = r'C:\Users\HalleyandWoody\Documents\Python\self-driving\P2-Advanced-Lane-Lines\output_images'

# List all images to iterate through
image_names = os.listdir("camera_cal/")


## In order to calibrate our camera, we need to find object pts and image pts
images = glob.glob('camera_cal\calibration*.jpg')

#objpoints are the same for all calibration images
objpoints = []
imgpoints = []
# The following two lines of code were taken from the video in lesson 5
objp = np.zeros((9*6,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

#loop through all images and find all imgpoints and objpoints
for imgpath in images:
    img = mpimg.imread(imgpath) # read in image
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #convert to grayscale
    
    # Find corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6), None)
    
    if ret == True:
        imgpoints.append(corners)
        objpoints.append(objp)
        
        #Visualize corners
        img = cv2.drawChessboardCorners(img, (9,6),corners, ret)
        plt.imshow(img)
    
#calibrate our camera given the objpoints and imgpoints just calculated
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)

#Undistort "calibration2.jpg" image in camera_cal folder and output the result to output_images folder
img = cv2.imread('camera_cal/calibration2.jpg')

undist = cv2.undistort(img, mtx, dist, None, mtx)
cv2.imwrite(os.path.join(path , 'Undistorted_Calibration_Image.jpg'), undist) # copies image into output_images directory


    
````

````python
### Apply sobel operators to create a thresholded binary image identifying lane lines

#  abs_sobel_thresh calculates threshold through directional gradient either x or y) 
def abs_sobel_thresh(img, orient='x', sobel_kernel=5, thresh=(50, 255)):
    # Calculate directional gradient
    # Apply threshold
        
    try:
        
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        sobel = cv2.Sobel(gray, cv2.CV_64F, orient == 'x', orient == 'y')

    except:
        
        sobel = img
        
        pass
    
        
    abs_sobel = np.absolute(sobel)
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    
    binary_output = np.zeros_like(scaled_sobel)
    binary_output[(scaled_sobel > thresh[0]) & (scaled_sobel < thresh[1])] = 1
    
    return binary_output

#  mag_thresh calculates threshold based on the magnitude of the gradient (calculated through magnitude x and magnitude y)
def mag_thresh(image, sobel_kernel=5, mag_thresh=(50, 255)):
        
    thresh_min = mag_thresh[0]
    thresh_max = mag_thresh[1]
    
        
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize = sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize = sobel_kernel)
        
    abs_sobel_xy = np.sqrt(np.square(sobelx) + np.square(sobely))
    
    scaled_sobel = np.uint8(255*abs_sobel_xy/np.max(abs_sobel_xy))
    binary_output = np.zeros_like(scaled_sobel)
    binary_output[(scaled_sobel > thresh_min) & (scaled_sobel < thresh_max)] = 1
    
    return binary_output
    
#  dir_thresh calculates threshold based on the angle of the gradient (calculated through magnitude x and magnitude y)
def dir_threshold(image, sobel_kernel=5, thresh=(0.7, 1.3)):
    
    thresh_min = thresh[0]
    thresh_max = thresh[1]
    
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize = sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize = sobel_kernel)
        
    # abs_sobel_xy = np.sqrt(np.square(sobelx) + np.square(sobely))
    abs_sobelx = np.absolute(sobelx)
    abs_sobely = np.absolute(sobely)
    
    direction_sobel = np.arctan2(abs_sobely, abs_sobelx)
    
    scaled_sobel = direction_sobel
    binary_output = np.zeros_like(scaled_sobel)
    binary_output[(scaled_sobel > thresh_min) & (scaled_sobel < thresh_max)] = 1

    return binary_output


# Sobel kernel size
ksize = 7 
image1 = undistorted_images[2]
# Apply each of the thresholding functions
gradx = abs_sobel_thresh(image1, orient='x', sobel_kernel=ksize, thresh=(50, 255))
grady = abs_sobel_thresh(image1, orient='y', sobel_kernel=ksize, thresh=(50, 255))
mag_binary = mag_thresh(image1, sobel_kernel=ksize, mag_thresh=(50, 255))
dir_binary = dir_threshold(image1, sobel_kernel=ksize, thresh=(0.7, 1.3))

combined_gradient1 = np.zeros_like(dir_binary)
combined_gradient1[((gradx == 1) & (grady == 1)) | ((mag_binary == 1) & (dir_binary == 1))] = 1

# print(combined)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image1)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(combined_gradient1, cmap='gray')
ax2.set_title('Combined Gradient', fontsize=50)

combined_out = 255*combined_gradient1 # convert the image to 8 bit for the cv2.imwrite() function
cv2.imwrite(os.path.join(path , 'Combined_Gradient.jpg'), combined_out) # copies image into output_images directory
````

````python
### Apply color transforms, notably on the S channel to separate lanes and combine with
### sobel operators from above to highlight the entire lane line as opposed to only the edges. Also, color transforms 
### Are needed in order to account for frames where the road is light in color and the edge detection algorithm
### cannot always detect those differences (image examples "test1" "test 4" and "test 5")

# Define function that takes as input a RGB image and outputs a binary thresholded image based on the Saturation and 
# lightness channels

def hls_saturation(img, s_thresh=(100, 255), l_thresh=(40, 255)):
    
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    S = hls[:,:,2]
    L = hls[:,:,1]
    binary_output = np.zeros_like(S)
    binary_output[(S > s_thresh[0]) & (S <= s_thresh[1]) & (L > l_thresh[0]) & (L <= l_thresh[1])] = 1

    return binary_output

combined_color = hls_saturation(image1, s_thresh=(100,255), l_thresh=(40,255))

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image1)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(combined_color, cmap='gray')
ax2.set_title('Combined Color', fontsize=50)

combined_out = 255*combined_color # convert the image to 8 bit for the cv2.imwrite() function
cv2.imwrite(os.path.join(path , 'Combined_Color.jpg'), combined_out) # copies image into output_images directory
````

````python
### Combining color and gradient thresholds

def Combined_colorgradient(img):
    
    # find the combined gradient binary first the same way as explained above 
    gradx = abs_sobel_thresh(img, orient='x', thresh=(100, 255))
    grady = abs_sobel_thresh(img, orient='y', thresh=(100, 255))
    mag_binary = mag_thresh(img, sobel_kernel=ksize, mag_thresh=(50, 255))
    dir_binary = dir_threshold(img, sobel_kernel=ksize, thresh=(0.7, 1.3))

    combined_gradient = np.zeros_like(dir_binary)
    combined_gradient[((gradx == 1) & (grady == 1)) | ((mag_binary == 1) & (dir_binary == 1))] = 1
    
    # find the combined color binary to combine with gradient binary and achieve more reliable results
    combined_color = hls_saturation(img, s_thresh=(100,255), l_thresh=(40,255))
    
    # combine both binaries
    combined_color_gradient = np.zeros_like(combined_gradient)
    combined_color_gradient[(combined_color == 1) | (combined_gradient == 1)] = 1

    return combined_color_gradient

combined_colorgradient = Combined_colorgradient(image1)

combined_out = 255*combined_colorgradient # convert the image to 8 bit for the cv2.imwrite() function
cv2.imwrite(os.path.join(path , 'Color_and_gradient_thresholded.jpg'), combined_out) # copies image into output_images directory

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image1)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(combined_colorgradient, cmap='gray')
ax2.set_title('Combined Color + Gradient', fontsize=50)
````

````python
### Sliding windows approach

# Following function takes in a warped binary image and returns x and y coordinated for the pixels constituting the 
# lane lines fir both left and right lane. This function also returns a visualization of the input image with slidinw windows
def find_lane_pixels(binary_warped):
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50

    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),
        (win_xleft_high,win_y_high),(0,255,0), 5) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),
        (win_xright_high,win_y_high),(0,255,0), 5) 
        
        # Identify the nonzero pixels in x and y within the window #
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        # If we found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]

    return leftx, lefty, rightx, righty, out_img


## The following function fits a second order polynomial on a warped binary image. It implements the function 
## find_lane_pixels above to get the lane pixels then uses numpy's polyfit to fit a 2nd order polynomial to those pixels
def fit_polynomial(binary_warped):
    # Find our lane pixels first
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)
    ### Fit a second order polynomial
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)
    
#     print(left_fit)

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty

    ## Visualization ##
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]

    # Plots the left and right polynomials on the lane lines
    plt.plot(left_fitx, ploty, color='yellow')
    plt.plot(right_fitx, ploty, color='yellow')

    return out_img, ploty, left_fit, right_fit


out_img, ploty, left_fit, right_fit = fit_polynomial(warped)

plt.imshow(out_img)

# using pylab to save the figure as opposed to saving the image since if I save the image the fit lines wont show
pylab.savefig('output_images/Sliding_Window_polynomial.jpg')
pylab.savefig('Sliding_Window_polynomial.jpg', bbox_inches='tight')
````

````python
### Calculate radius of curvature in useful real world units (meters)

# Convert pixel curve to useful real world units (meters) in the following function
def measure_curvature_real(binary_warped):

    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
        
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)
        
    left_fit_cr = np.polyfit(ym_per_pix*lefty, xm_per_pix*leftx, 2)
    right_fit_cr = np.polyfit(ym_per_pix*righty, xm_per_pix*rightx, 2)
    

    
    # Define y-value where we want radius of curvature 
    y_eval = np.max(ploty) #pick y at the bottom of the image
    
    left_curverad = ((1+(2*left_fit_cr[0]*y_eval+left_fit_cr[1])**2)**(1.5))/np.absolute(2*left_fit_cr[0])  ## calculation of the left line
    right_curverad = ((1+(2*right_fit_cr[0]*y_eval+right_fit_cr[1])**2)**(1.5))/np.absolute(2*right_fit_cr[0])  ## calculation of the right line
    
    return left_curverad, right_curverad

# Calculate the radius of curvature in meters for both lane lines
left_curverad, right_curverad = measure_curvature_real(warped)

print(left_curverad, 'm', right_curverad, 'm')
````

````python
### detecting vehicle offset from center of lane in meters 

def vehicle_offset(binary_warped):
    
    # Define conversions in x from pixels space to meters
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    
#     # finding lane pixel parameters in order to feed those into fit_poly function
#     leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)
#     # finding polynomial fit lines that will help us find vehicle offset (found in units of meters)
#     left_fit_cr = np.polyfit(ym_per_pix*lefty, xm_per_pix*leftx, 2)
#     right_fit_cr = np.polyfit(ym_per_pix*righty, xm_per_pix*rightx, 2)
    
    # finding lane pixel parameters in order to feed those into np.polyfit function
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)
    # finding polynomial fit lines that will help us find vehicle offset (found in units of meters)
    left_fit_cr = np.polyfit(lefty, leftx, 2)
    right_fit_cr = np.polyfit(righty, rightx, 2)    
    
    #Calculate the x-value of the polynomial fit at the bottom of the image for both left and right fits
    bottom_left_fitx = left_fit_cr[0]*binary_warped.shape[0]**2 + left_fit_cr[1]*binary_warped.shape[0] + left_fit_cr[2]
    bottom_right_fitx = right_fit_cr[0]*binary_warped.shape[0]**2 + right_fit_cr[1]*binary_warped.shape[0] + right_fit_cr[2]
    
    # find offset by subtracting the following two values
    
    camera_mid_x = binary_warped.shape[1]/2
    lane_mid_x = (bottom_left_fitx + bottom_right_fitx)/2
    
    diff = camera_mid_x - lane_mid_x
    
    # convert to meters
    diff_meters = diff*xm_per_pix
    
    return diff_meters

# Test function, result can be negative (meaning the car is to the left of the center) or positive (car is to the right)
# result is in meters
print(vehicle_offset(warped))
    
````

````python
### Transform the image back to the original view and add radius of curvature and distance from center

# plt.imshow(unwarped, cmap = 'gray')

def unwarp_image(binary_warped, undistorted):
        
    # Create an image to draw the lines on
    warp_zero = np.zeros_like(binary_warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    # find lane pixels
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)
    # find polynomial fits
    left_fitx, right_fitx, ploty = fit_poly(binary_warped.shape, leftx, lefty, rightx, righty)

    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    
    
    transf, src, dst = perspective_transform(binary_warped)
    
    Minv = cv2.getPerspectiveTransform(dst, src)
    
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (binary_warped.shape[1], binary_warped.shape[0])) 
    # Combine the result with the original image
    result = cv2.addWeighted(undistorted, 1, newwarp, 0.3, 0)    
    
    
    # find curvature and write values down on image
    curve1, curve2 = measure_curvature_real(binary_warped) # find left and right curvatures
    curve = (curve1 + curve2)/2

    # find distance from center and write values down on image
    offset = vehicle_offset(binary_warped)
    check = ('left', 'right')
    pos = check[offset > 0] # determines if we should write 'left' or 'right' of center

    cv2.putText(result, 'Radius of Curvature = ' + str(int(curve)) + '(m)', (50, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, lineType=cv2.LINE_AA)
    cv2.putText(result, 'Vehicle is ' + str(abs(offset))[:4] + 'm ' + pos + ' of center', (50, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, lineType=cv2.LINE_AA)
    
    return result


result =  unwarp_image(warped, image1)

# # find curvature and write values down on image
# curve1, curve2 = measure_curvature_real(warped) # find left and right curvatures
# curve = (curve1 + curve2)/2

# # find distance from center and write values down on image
# offset = vehicle_offset(warped)
# check = ('left', 'right')
# pos = check[offset > 0] # determines if we should write 'left' or 'right' of center

# cv2.putText(result, 'Radius of Curvature = ' + str(int(curve)) + '(m)', (50, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, lineType=cv2.LINE_AA)
# cv2.putText(result, 'Vehicle is ' + str(abs(offset))[:4] + 'm ' + pos + ' of center', (50, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, lineType=cv2.LINE_AA)

plt.imshow(result)


image_write = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
cv2.imwrite(os.path.join(path , 'identified_lane_image.jpg'), image_write) # copies image into output_images directory
````
